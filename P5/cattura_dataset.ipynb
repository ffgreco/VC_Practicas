```python
import cv2
import os


CLASS_NAME = "smile"   

OUTPUT_DIR = os.path.join("dataset", CLASS_NAME)
os.makedirs(OUTPUT_DIR, exist_ok=True)

count = 0

print("Premi la BARRA SPAZIATRICE per salvare una foto")
print("Premi 'q' nel riquadro video per uscire")
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    cv2.imshow("Cattura dataset - " + CLASS_NAME, frame)
    key = cv2.waitKey(1) & 0xFF
    
    if key == ord(' '):   # Barra spaziatrice
        filename = os.path.join(OUTPUT_DIR, f"{CLASS_NAME}_{count:03d}.jpg")
        cv2.imwrite(filename, frame)
        print("Salvata:", filename)
        count += 1
        
    elif key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import os
import cv2
import numpy as np
from skimage.feature import hog
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

DATASET_DIR = "dataset"

classes = sorted([
    d for d in os.listdir(DATASET_DIR)
    if os.path.isdir(os.path.join(DATASET_DIR, d))
])

print("Classi trovate:", classes)

X = []  # feature HOG
Y = []  # etichette

for label in classes:
    folder = os.path.join(DATASET_DIR, label)
    for fname in os.listdir(folder):
        if not fname.lower().endswith((".jpg", ".jpeg", ".png")):
            continue

        img_path = os.path.join(folder, fname)
      
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            print("Errore nel leggere:", img_path)
            continue

       
        img = cv2.resize(img, (128, 128))

       
        features = hog(
            img,
            pixels_per_cell=(16, 16),
            cells_per_block=(2, 2),
            block_norm='L2-Hys'
        )

        X.append(features)
        Y.append(label)


X = np.array(X)
Y = np.array(Y)

print("Shape X:", X.shape)
print("Etichette e conteggi:", np.unique(Y, return_counts=True))


X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.3, random_state=42, stratify=Y
)


clf = SVC(kernel="linear")
clf.fit(X_train, Y_train)

Y_pred = clf.predict(X_test)

print("\nREPORT:\n")
print(classification_report(Y_test, Y_pred))
print("Matrice di confusione:")
print(confusion_matrix(Y_test, Y_pred))
```
